```python
import requests
from bs4 import BeautifulSoup
import csv
import random, time

url_a = 'https://www.104.com.tw/jobs/search/?ro=0&keyword=後端&expansionType=area,spec,com,job,wf,wktm&order=12&asc=0&page='
url_b = '&mode=s&jobsource=2018indexpoc&langFlag=0&langStatus=0&recommendJob=1&hotJob=1'

all_job_datas = []

for page in range(1,3+1):
    url = url_a+ str(page)+url_b
    print(url)
    headers = {"Referer": "https://www.104.com.tw/job/6vrlf"}
    response = requests.get(url, headers =headers)
    soup = BeautifulSoup(response.text, 'html.parser')
    jobs = soup.find_all('article',class_='js-job-item')
    
    for job in jobs:
        job_name = job.find('a', class_='js-job-link').text
        job_company = job.get('data-cust-name')
        job_loc = job.find('ul', class_='job-list-intro').find('li').text
        job_pay = job.find('span',class_='b-tag--default').text
        
        job_data = {'職缺':job_name, '公司名稱':job_company, '地址':job_loc, '薪資':job_pay}
        
        all_job_datas.append(job_data)
        time.sleep(random.randint(1,3))

# fn = '104後端.csv'
# columns_name = ['職缺','公司','地址', '薪資',]
# with open(fn, 'w', newline='') as f:
#     dictWriter = csv.DictWriter(f, fieldnames = columns_name)
#     dictWriter.writeheader()
#     for data in all_job_datas:
#         dictWriter.writerow(data)
```

```python
import requests
import re
     
# 取得免費IP
#source1: https://www.learncodewithmike.com/2021/10/python-scrape-free-proxy-ip.html
#source2: https://blog.csdn.net/qq_40727946/article/details/103731840
#source3: https://www.learncodewithmike.com/2021/10/python-scraper-rotate-and-random-proxy-ip.html
response = requests.get("https://www.sslproxies.org/")
proxy_ips = re.findall('\d+\.\d+\.\d+\.\d+:\d+', response.text)  #「\d+」代表數字一個位數以上
# print(proxy_ips)

valid_ips = []
for ip in proxy_ips:
    try:
        result = requests.get('https://ip.seeip.org/jsonip?',
                             proxies={'http':ip, 'https':ip},
                             timeout=5)
        print(result.json())
        valid_ips.append(ip)
    except:
        print(f'{ip} invalid')
```

```python
with open('proxy_list.text','w') as f1:
    for ip in valid_ips:
        f1.write('http://'+ip + '\n')
    f1.close()
```

```python
import random
print(random.choices(valid_ips))
```

```python
import random

def get_random_proxy():
    # fr = open(r'proxy_list.text').read()
    proxypool_url = random.choices(valid_ips)
    return proxypool_url
```

```python
import requests
from bs4 import BeautifulSoup
import csv
import random, time
from fake_useragent import UserAgent

delay_choices = [1,2,3,4]#延遲秒數
delay = random.choices(delay_choices)#隨機選取秒數

ua = UserAgent()
# proxy = get_random_proxy()
# proxies1 = 'https://' + proxy[0]
# proxies2 = 'http://' + proxy[0]
# proxies3 = {'http': proxies2, 'https':proxies1}

url_a = 'https://www.104.com.tw/jobs/search/?ro=0&keyword=前端&expansionType=area,spec,com,job,wf,wktm&order=12&asc=0&page='
url_b = '&mode=s&jobsource=2018indexpoc&langFlag=0&langStatus=0&recommendJob=1&hotJob=1'

all_job_datas = []

for page in range(1,151):
    url = url_a+ str(page)+url_b
    print(url)
    headers = {'Referer': 'https://www.104.com.tw/job/6vrlf',
              'User-Agent': ua.random}
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.text, 'html.parser')
    jobs = soup.find_all('article',class_='js-job-item')
    
    for job in jobs:
        job_name = job.find('a', class_='js-job-link').text
        job_company = job.get('data-cust-name')
        job_loc = job.find('ul', class_='job-list-intro').find('li').text
        job_pay = job.find('span',class_='b-tag--default').text
        job_content_link = job.find('a', class_='js-job-link').get('href')
        all_job_link = ('https:'+job_content_link)
        
        print('抓職缺大數據中...')
        
        header = {'Referer': 'https://www.104.com.tw/jobs/search/'}
        time.sleep(1)
        response_link = requests.get(all_job_link, headers = header)
        time.sleep(1)
        soup_link = BeautifulSoup(response_link.text, 'html.parser')
        try:
            job_content = soup_link.find('p', class_='job-description__content').text #工作內容
        except:
            job_content = ''
        job_require_all = soup_link.find_all('div', class_= 'dialog container-fluid bg-white rounded job-requirement mb-4 pt-6 pb-6')
        job_goodat_all = soup_link.find_all('a', class_='tools')
        job_other_all = soup_link.find_all('p', class_='m-0 r3 w-100')
        
        job_goodat1 = []
        list.clear(job_goodat1)
        
        # for link in all_job_link:
        #     url = link
        #     header = {'Referer': 'https://www.104.com.tw/jobs/search/'}
        #     response_link = requests.get(url, headers = header)
        #     time.sleep(1)
        #     soup_link = BeautifulSoup(response_link.text, 'html.parser')
        #     job_content = soup_link.find('p', class_='job-description__content').text #工作內容
        #     job_require_all = soup_link.find_all('div', class_= 'dialog container-fluid bg-white rounded job-requirement mb-4 pt-6 pb-6')
        #     job_goodat_all = soup_link.find_all('a', class_='tools')
        #     job_other_all = soup_link.find_all('p', class_='m-0 r3 w-100')
        #     print('抓detail中...')
            
        for i in job_require_all:
            job_exp = i.p.text.strip()#工作經歷
            job_dreqire = i.find_all('p')[1].text.strip()#學歷要求
            job_mreqire = i.find_all('p')[2].text.strip()#科系要求
        for i in job_goodat_all:
            job_goodat = i.text #擅長工具
            job_goodat1.append(job_goodat)
        for i in job_other_all:
            job_other = i.text # 其他條件
                
        print('寫入細節')    
        job_data = {'職缺':job_name, 
                    '公司名稱':job_company, 
                    '地址':job_loc, 
                    '薪資':job_pay, 
                    '連結':job_content_link, 
                    '工作內容':job_content, 
                    '工作經驗':job_dreqire, 
                    '科系要求':job_mreqire, 
                    '擅長工具':job_goodat1, 
                    '其他條件':job_other}
        
        all_job_datas.append(job_data)
        print('資料寫入完成')
        time.sleep(delay[0])       
    time.sleep(delay[0])
```
